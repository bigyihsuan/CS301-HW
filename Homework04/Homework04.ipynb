```python
import pandas as pd
import numpy as np
def data_reader(path):
    try:
        data = pd.read_csv(path)
    except:
        print("Error during reading, please try again.")
        return None
    return data
```

# Problem 1

```python
def get_percent_nans(df, column_name):
    return len(df.loc[(df[column_name].isna())]) / len(df) * 100
```

```python
data = data_reader('2019_nCoV_20200121_20200206.csv')
print(get_percent_nans(data, "Suspected"))
```

```python
def get_continent(country):
    countries = {
        'Mainland China':'Asia','Singapore':'Asia','Thailand':'Asia','Japan':'Asia','Hong Kong':'Asia','South Korea':'Asia',
        'Germany':'Europe','Malaysia':'Asia','Taiwan':'Asia','Macau':'Asia','Vietnam':'Asia','France':'Europe',
        'United Arab Emirates':'Asia','Australia':'Australia','India':'Asia','Canada':'North America','Italy':'Europe',
        'Philippines':'Asia','Russia':'Europe','UK':'Europe','United States':'North America','Belgium':'Europe',
        'Cambodia':'Asia','Finland':'Europe','Nepal':'Asia','Spain':'Europe','Sri Lanka':'Asia','Sweden':'Europe',
        'Ivory Coast':'Africa','Mexico':'North America','Brazil':'South America','Colombia':'South America'
    }
    return countries[country]
print(get_continent("UK"))
print(get_continent("Ivory Coast"))
```

# Problem 7

```python
continents = data["Country/Region"].apply(get_continent, (data["Country/Region"].item))
continents.name = "Continent"
continents
```

```python
data = data.join(continents)
data
```
